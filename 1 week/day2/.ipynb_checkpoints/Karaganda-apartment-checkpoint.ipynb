# Караганда — квартиры

```python
import pandas as pd
```

```python
df = pd.read_csv("karaganda_apartments.csv")
```

```python
# Размер таблицы: (строки, столбцы)
df.shape
```

```python
# Первые 10 строк
df.head(10)
```

```python
# Типы данных и общая информация
df.info()
```

```python
# Количество пропусков по столбцам
df.isnull().sum()
```

```python
# Доля пропусков в процентах
(df.isnull().mean() * 100).round(1)
```

```python
# Количество полных дубликатов строк
df.duplicated().sum()
```

```python
# Количество уникальных значений по столбцам
df.nunique()
```

```python
# Уникальные значения категориальных признаков
for col in ["rooms", "balcony", "parking", "furnished", "ceiling_height"]:
    print(col, "->", df[col].unique())
```

```python
# Базовая статистика по числовым столбцам
df.describe()
```

```python
# Статистика по всем столбцам, включая текстовые
df.describe(include="all")
```
