## 20 простых вопросов для исследования данных

```python
#  Самая высокая цена
df["PRICE"].max()
```

```python
#  Самая низкая цена
df["PRICE"].min()
```

```python
#  Топ-10 районов по числу объявлений
df["SUBLOCALITY"].value_counts().head(10)
```

```python
#  Средняя цена по районам
df.groupby("SUBLOCALITY")["PRICE"].mean()
```

```python
#  Сколько домов дороже 1 млн (фильтр)
(df["PRICE"] > 1_000_000).sum()
```

```python
# 11. Средняя площадь (кв. футы)
df["PROPERTYSQFT"].mean()
```

```python
# 12. Самый большой объект по площади
df.loc[df["PROPERTYSQFT"].idxmax(), ["PROPERTYSQFT", "PRICE", "ADDRESS"]]
```

```python
# 13. Средняя цена за кв. фут
(df["PRICE"] / df["PROPERTYSQFT"]).mean()
```

```python
# 14. Сколько объектов по числу ванных
df["BATH"].value_counts()
```

```python
# 15. Топ-10 брокеров по числу объявлений
df["BROKERTITLE"].value_counts().head(10)
```

```python
# 16. Средняя цена по числу спален
df.groupby("BEDS")["PRICE"].mean()
```

```python
# 17. Связь цены, спален, ванных и площади (корреляция)
df[["PRICE", "BEDS", "BATH", "PROPERTYSQFT"]].corr()
```

```python
# 18. Сколько уникальных городов (LOCALITY)
df["LOCALITY"].nunique()
```

```python
# 19. Дома с 3 спальнями дешевле 500 тыс (комбинированный фильтр)
df[(df["BEDS"] == 3) & (df["PRICE"] < 500_000)]
```

```python
# 20. Средняя цена и площадь по типу объекта
df.groupby("TYPE")[["PRICE", "PROPERTYSQFT"]].mean()
```
