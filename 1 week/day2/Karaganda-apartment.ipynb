```python
print("Максимум:", df["area_m2"].max())
print("Минимум:", df["area_m2"].min())
df.nlargest(5, "price_tenge")
```

**Что заметил:** есть квартира 14 м² и квартира 340 м² за 190 млн — крайности, которые тянут среднее.
**Как проверяю:** найду выбросы по цене формально — через межквартильный размах (правило `1.5·IQR`). Это те объекты, которые стоит проверять глазами и, возможно, исключать из моделей.

```python
q1, q3 = df["price_tenge"].quantile([0.25, 0.75])
iqr = q3 - q1
low, high = q1 - 1.5 * iqr, q3 + 1.5 * iqr
print(f"Нормальный диапазон цены: {low:,.0f} — {high:,.0f} тг")

outliers = df[(df["price_tenge"] < low) | (df["price_tenge"] > high)]
print("Выбросов по цене:", len(outliers))
outliers[["rooms", "area_m2", "price_tenge", "price_per_m2", "address"]].sort_values("price_tenge", ascending=False)
```

## Проверка адекватности: м² на комнату

**Зачем:** хороший способ поймать «битые» строки — площадь на одну комнату. Слишком мало (< 12 м²) или слишком много (> 60 м²) — повод присмотреться: ошибка в данных или нетипичный объект (студия / элитка).

```python
df["m2_per_room"] = (df["area_m2"] / df["rooms"]).round(1)
print("Самые «тесные» (мало м² на комнату):")
print(df.nsmallest(5, "m2_per_room")[["title", "rooms", "area_m2", "m2_per_room", "price_tenge"]])
```

```python
print("\nСамые «просторные» на комнату:")
df.nlargest(5, "m2_per_room")[["title", "rooms", "area_m2", "m2_per_room", "price_tenge"]]
```
