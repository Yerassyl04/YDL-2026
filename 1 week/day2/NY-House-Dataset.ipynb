## 20 простых вопросов для исследования данных

```python
#  Самая высокая цена
df["PRICE"].max()
```

```python
#  Самая низкая цена
df["PRICE"].min()
```

```python
#  Топ-10 районов по числу объявлений
df["SUBLOCALITY"].value_counts().head(10)
```

```python
#  Средняя цена по районам
df.groupby("SUBLOCALITY")["PRICE"].mean()
```

```python
#  Сколько домов дороже 1 млн (фильтр)
(df["PRICE"] > 1_000_000).sum()
```

```python
#  Средняя площадь (кв. футы)
df["PROPERTYSQFT"].mean()
```

```python
#  Самый большой объект по площади
df.loc[df["PROPERTYSQFT"].idxmax(), ["PROPERTYSQFT", "PRICE", "ADDRESS"]]
```

```python
#  Средняя цена за кв. фут
(df["PRICE"] / df["PROPERTYSQFT"]).mean()
```

```python
# Средняя цена по числу спален
df.groupby("BEDS")["PRICE"].mean()
```

```python
#  Связь цены, спален, ванных и площади
df[["PRICE", "BEDS", "BATH", "PROPERTYSQFT"]].corr()
```

```python
# Сколько уникальных городов (LOCALITY)
df["LOCALITY"].nunique()
```

```python
#  Дома с 3 спальнями дешевле 500 тыс 
df[(df["BEDS"] == 3) & (df["PRICE"] < 500_000)]
```

```python
# Средняя цена и площадь по типу объекта
df.groupby("TYPE")[["PRICE", "PROPERTYSQFT"]].mean()
```

```python
#  В каком районе жильё в среднем самое дорогое?
df.groupby("SUBLOCALITY")["PRICE"].median().sort_values(ascending=False).head(10)
```

```python
#  Какой тип жилья самый дорогой, а какой доступный?
df.groupby("TYPE")["PRICE"].median().sort_values(ascending=False)
```

```python
#  Как растёт цена с числом спален?
df.groupby("BEDS")["PRICE"].median().head(8)
```

```python
#  В каких районах больше всего предложений на рынке?
df["SUBLOCALITY"].value_counts().head(10)
```

```python
# Средняя цена за квадратный фут по районам
df["price_per_sqft"] = df["PRICE"] / df["PROPERTYSQFT"]
df.groupby("SUBLOCALITY")["price_per_sqft"].mean().sort_values(ascending=False).head(10)
```

## Краткое саммари

**Объём данных:** 4801 строка, 17 столбцов. Пропусков нет, но есть **214 полных дубликатов** (стоит удалить через `df.drop_duplicates()`).

**Основные выводы:**
- **Цены** сильно разбросаны: медиана ≈ 825 тыс., среднее ≈ 2.36 млн — распределение скошено вправо из-за дорогих объектов.
- **Самые дорогие районы** (по медиане): New York County (~6.5 млн), Dumbo (~5.8 млн), Manhattan (~2.8 млн). Самые доступные — Queens, The Bronx, Staten Island.
- **По типу жилья**: дороже всего Townhouse (медиана ~2.95 млн), доступнее всего Co-op (~425 тыс.).
- **Цена растёт с числом спален** (медиана от 395 тыс. за 1 спальню до 1.87 млн за 8), но связь нелинейная.
- **Цена за кв. фут** максимальна в Dumbo, New York County и Manhattan (~1800–2300 $/sqft).
- **Корреляции** PRICE с BEDS/BATH/PROPERTYSQFT слабые (0.05–0.11) — в основном из-за выбросов.
- **Больше всего предложений** в New York, Kings County, Queens County.

**⚠️ Замечания по качеству данных (видны из запросов):**
- `PRICE` max = **2 147 483 647** (это максимум int32) — явно ошибочное/заглушечное значение, искажает среднее.
- `PROPERTYSQFT` = **2184.207862** многократно повторяется — похоже на значение-заполнитель (импутация средним).
- `BATH` содержит дробные значения вроде **2.373861** (то же среднее) — тоже признак заполнения пропусков.
- Категории районов дублируются по смыслу (`New York`/`New York County`/`Manhattan`, `Kings County`/`Brooklyn`) — стоит унифицировать перед группировкой.
