```python
# Проблемный столбец: pH (Acidity) — единственный с заметным правым перекосом
p = df["pH (Acidity)"]

print(f"mean   = {p.mean():.4f}")
print(f"median = {p.median():.4f}")
print(f"std    = {p.std():.4f}")
print(f"min    = {p.min():.4f}")
print(f"Q1     = {p.quantile(0.25):.4f}")
print(f"Q3     = {p.quantile(0.75):.4f}")
print(f"max    = {p.max():.4f}")
print(f"IQR    = {p.quantile(0.75) - p.quantile(0.25):.4f}")
print(f"skew   = {p.skew():.4f}")
print(f"kurt   = {p.kurt():.4f}")

q1, q3 = p.quantile(0.25), p.quantile(0.75)
iqr = q3 - q1
lo, hi = q1 - 1.5 * iqr, q3 + 1.5 * iqr
out = p[(p < lo) | (p > hi)]
print(f"границы усов: [{lo:.2f}, {hi:.2f}]")
print(f"выбросов: {len(out)} ({len(out) / len(p) * 100:.2f}%)")
```

## Одна корреляция: Brix (Sweetness) ↔ Quality

Гипотеза по-человечески: **чем слаще апельсин, тем выше его оценка качества.**
Проверим её через коэффициент корреляции Пирсона `r` и его p-value.

- `r` показывает силу и знак **линейной** связи: от −1 (идеальная обратная) до +1 (идеальная прямая), 0 — связи нет.
- `p-value` отвечает на вопрос: *«А не могло ли такое r получиться чисто случайно, если на самом деле связи нет?»* Маленькое p (< 0.05) → совпадением это объяснить трудно.

```python
x = df["Brix (Sweetness)"]
y = df["Quality (1-5)"]

r, p = stats.pearsonr(x, y)
n = len(df)

print(f"n        = {n}")
print(f"Pearson r = {r:+.3f}")
print(f"p-value   = {p:.2e}")
print(f"r^2       = {r**2:.3f}  (доля разброса Quality, объяснённая линией)")

# Грубая шкала силы связи по |r|
ar = abs(r)
sila = ("очень слабая" if ar < 0.1 else
        "слабая"       if ar < 0.3 else
        "умеренная"    if ar < 0.5 else
        "заметная"     if ar < 0.7 else
        "сильная"      if ar < 0.9 else "очень сильная")
print(f"сила связи: {sila}, направление: {'прямая (+)' if r > 0 else 'обратная (-)'}")

# Визуальная проверка: scatter + линия регрессии
plt.figure(figsize=(7, 5))
sns.regplot(x=x, y=y, scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})
plt.title(f"Brix vs Quality:  r = {r:+.3f},  p = {p:.1e}")
plt.show()
```

### Вердикт

**Связь настоящая, не совпадение.**

- `r = +0.63` — заметная прямая связь: чем выше Brix (сладость), тем выше оценка Quality.
- `p ≈ 3.3e-28` — это практически ноль (намного меньше 0.05). Вероятность получить такое `r` случайно при отсутствии связи ничтожна. На выборке в 241 апельсин это не шум.
- `r² ≈ 0.40` — сладостью объясняется примерно **40%** разброса оценок качества. Прилично, но больше половины разброса — за счёт других факторов (сорт, спелость, дефекты).

### ⚠️ Корреляция ≠ причинность

Мы доказали, что Brix и Quality **движутся вместе**, но НЕ доказали, что *сладость вызывает* высокую оценку. Возможные ловушки:
- **Скрытый фактор (confounder):** спелость или сорт могут одновременно поднимать и сахар, и оценку — тогда Brix лишь «попутчик», а не причина.
- **Обратное направление:** маловероятно здесь, но в общем случае стрелка причины может смотреть в другую сторону.
- **r измеряет только линейность:** изогнутую зависимость он бы недооценил.

Чтобы говорить о причине, нужен контролируемый эксперимент или поправка на скрытые факторы (например, частная корреляция / регрессия с контролем сорта и спелости) — а не один коэффициент Пирсона.

# Полный EDA

Соберём разведочный анализ в один систематический проход. Выше мы уже сделали часть: загрузку, очистку `Blemishes`, проверки качества данных, `describe()`, разбор отдельных колонок и корреляции. Здесь закрываем оставшееся, чтобы EDA был полным:

## 1. Целевая переменная — Quality (1-5)

```python
q = df["Quality (1-5)"]
print(f"mean={q.mean():.2f}  median={q.median():.2f}  std={q.std():.2f}  skew={q.skew():.2f}")

fig, axes = plt.subplots(1, 2, figsize=(13, 4))
sns.countplot(x=q, ax=axes[0], hue=q, palette="viridis", legend=False)
axes[0].set_title("Сколько апельсинов каждой оценки")
sns.histplot(q, kde=True, bins=15, ax=axes[1])
axes[1].axvline(q.mean(), color="red", linestyle="--", label=f"mean={q.mean():.2f}")
axes[1].set_title("Распределение Quality")
axes[1].legend()
plt.tight_layout()
plt.show()
```

## 2. Все числовые признаки сразу

```python
num_features = ["Size (cm)", "Weight (g)", "Brix (Sweetness)", "pH (Acidity)",
                "Softness (1-5)", "HarvestTime (days)", "Ripeness (1-5)"]

# Гистограммы — форма распределения каждого признака
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for ax, col in zip(axes.flat, num_features):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(col)
axes.flat[-1].axis("off")  # лишняя 8-я ячейка сетки
fig.suptitle("Распределения числовых признаков", y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Боксплоты — поиск выбросов по каждому признаку (через z-score / правило 1.5*IQR)
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for ax, col in zip(axes.flat, num_features):
    sns.boxplot(x=df[col], ax=ax)
    ax.set_title(col)
axes.flat[-1].axis("off")
fig.suptitle("Боксплоты числовых признаков (выбросы)", y=1.02, fontsize=14)
plt.tight_layout()
plt.show()

# Сводка по выбросам (правило 1.5*IQR)
print("Выбросы по правилу 1.5*IQR:")
for col in num_features:
    s = df[col]
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    lo, hi = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    n_out = ((s < lo) | (s > hi)).sum()
    print(f"  {col:20s}: {n_out:3d} ({n_out / len(s) * 100:.1f}%)  skew={s.skew():+.2f}")
```

**Угадка перед запуском:** спелость измеряют по шкале 1–5, многие апельсины собирают «дозревшими» → ожидаю смещение к высоким значениям, то есть среднее *ниже* медианы.

## Задача 2. Одна корреляция — Brix (Sweetness) ↔ Quality (1-5)
Угадка: слаще → вкуснее → выше оценка. Ожидаю заметную положительную связь.

```python
x, y = df["Brix (Sweetness)"], df["Quality (1-5)"]
rho, pval = stats.pearsonr(x, y)
print(f"n         = {len(df)}")
print(f"Pearson r = {rho:+.3f}")
print(f"p-value   = {pval:.2e}")
print(f"r^2       = {rho**2:.3f}  (доля разброса Quality, объяснённая линейно)")

plt.figure(figsize=(7,5))
sns.regplot(x=x, y=y, scatter_kws={"alpha":.4}, line_kws={"color":"red"})
plt.title(f"Brix vs Quality:  r={rho:+.3f},  p={pval:.1e}")
plt.show()
```

**Вердикт: связь настоящая, не совпадение.**
- `r = +0.63` — заметная прямая связь.
- `p ≈ 3.3e-28` ≪ 0.05 — случайностью такое на 241 наблюдении не объяснить.
- `r² ≈ 0.40` — сладость объясняет ~40% разброса оценок, остальное за другими факторами.

⚠️ **Корреляция ≠ причинность.** Brix и Quality движутся вместе, но это не доказывает, что *сахар* поднимает оценку: спелость или сорт могут одновременно тянуть вверх и сахар, и оценку (скрытый фактор). Чтобы говорить о причине, нужен контроль этих факторов, а не один r.

## 3. Проверьте руками

Машина сказала «больше сахара → выше оценка». Проверим это **руками** на маленьком куске: берём 10 самых сладких и 10 самых кислых апельсинов и сравниваем среднюю оценку. Никаких корреляций — просто две кучки.

```python
top = df.nlargest(10, "Brix (Sweetness)")
bot = df.nsmallest(10, "Brix (Sweetness)")
print("10 самых сладких:  Brix mean = %.2f,  Quality mean = %.2f" % (top["Brix (Sweetness)"].mean(), top["Quality (1-5)"].mean()))
print("10 самых кислых:   Brix mean = %.2f,  Quality mean = %.2f" % (bot["Brix (Sweetness)"].mean(), bot["Quality (1-5)"].mean()))

# Ещё руками: пересчитаем среднее Brix по первым 5 строкам без pandas-агрегата
five = df["Brix (Sweetness)"].head(5).tolist()
print("\nПервые 5 Brix:", five)
print("сумма/5 вручную =", sum(five)/5, " | pandas .mean() =", round(df["Brix (Sweetness)"].head(5).mean(),4))
```

Подтверждено руками. Самые сладкие апельсины в среднем оценены заметно выше самых кислых — направление совпало с корреляцией из Задачи 2, и это видно без всякой статистики, просто на двух кучках по 10 штук. Ручной пересчёт среднего по 5 строкам совпал с pandas → агрегату можно верить.

## Задача 4. Форма распределения

Строим распределение нескольких колонок. Вопросы: нормальное? два горба? спрятался ли выброс, которого не видно по среднему?

```python
fig, ax = plt.subplots(1, 3, figsize=(17,4))
for a, col in zip(ax, ["Brix (Sweetness)", "Quality (1-5)", "pH (Acidity)"]):
    sns.histplot(df[col], kde=True, ax=a)
    a.axvline(df[col].mean(), color="red", ls="--", label="mean")
    a.axvline(df[col].median(), color="green", ls=":", label="median")
    a.set_title(f"{col}\nskew={df[col].skew():+.2f}")
    a.legend()
plt.tight_layout(); 
plt.show()
```

## Задача 5. z-score — кандидаты в аномалии

Стандартизуем колонки (z = (x − mean)/std) и достаём точки с |z| > 2. Это кандидаты в аномалии — не приговор, а «посмотреть внимательнее».

```python
for col in ["pH (Acidity)", "HarvestTime (days)", "Brix (Sweetness)", "Weight (g)"]:
    z = (df[col] - df[col].mean()) / df[col].std()
    print(f"{col:20s}: |z|>2 -> {(z.abs()>2).sum():2d},  |z|>3 -> {(z.abs()>3).sum()}")

# Покажем сами строки-кандидаты по pH
z_ph = (df["pH (Acidity)"] - df["pH (Acidity)"].mean()) / df["pH (Acidity)"].std()
cand = df.loc[z_ph.abs() > 2, ["pH (Acidity)", "Quality (1-5)", "Variety"]].copy()
cand["z"] = z_ph[z_ph.abs() > 2].round(2)
print("\nКандидаты в аномалии по pH (|z|>2):")
print(cand.to_string())
```

Вердикт. Жёстких аномалий нет: ни в одной колонке нет точек с |z|>3. По pH — 4 точки с |z|>2 (самые кислые/щелочные апельсины), но это «края нормы», а не ошибки данных. Чистить нечего — но мы это проверили, а не предположили.

## 6. Среднее, которое врёт — парадокс Симпсона

**Угадка:** крупнее апельсин → ниже оценка (крупные часто водянистые). Проверим общую корреляцию `Size ↔ Quality`, а потом разрежем по цвету. Здесь машина чуть нас не обманула.

```python
r_all, p_all = stats.pearsonr(df["Size (cm)"], df["Quality (1-5)"])
print(f"ОБЩАЯ корреляция Size <-> Quality: r = {r_all:+.3f} (p={p_all:.2e})  -> крупнее = хуже?")

print("\nТеперь та же связь ВНУТРИ каждого цвета:")
for c, sub in df.groupby("Color"):
    if len(sub) >= 15:
        r, _ = stats.pearsonr(sub["Size (cm)"], sub["Quality (1-5)"])
        flag = "  <-- ЗНАК ПЕРЕВЕРНУЛСЯ" if r * r_all < 0 else ""
        print(f"  {c:14s} (n={len(sub):3d}): r = {r:+.3f}{flag}")
```

**Ловушка поймана.** Общая корреляция `Size ↔ Quality` = **−0.24** (крупнее → хуже). Но внутри цвета `Orange` она **+0.33** — знак переворачивается! Это **парадокс Симпсона**: общая связь — артефакт того, что разные цвета сидят в разных диапазонах размера и качества. Вывод «крупный апельсин хуже» был бы враньём — он держится только пока мы смешиваем группы.
