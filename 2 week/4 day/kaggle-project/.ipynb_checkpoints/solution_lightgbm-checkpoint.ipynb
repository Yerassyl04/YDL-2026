# Sleep-Stage Classification — Reference Solution

**Task.** Predict the sleep stage (`sleep_stage` ∈ {0,1,2,3}) from 21 physiological signals
(EEG / EMG / EOG / heart / respiration).
**Metric.** macro-F1 (every class weighs equally → errors on the hard class are expensive).

**Approach — kept deliberately simple and fast (runs in ~1 min on any laptop):**
1. **EDA that drives decisions** — class balance, missingness, feature signal, class separability,
   and a leakage/validation sanity check. (We skip slow decorative plots like t-SNE/UMAP.)
2. **Model: LightGBM** (gradient-boosted trees). For clean numeric tabular data this consistently
   beats RandomForest, trains in seconds, and **handles missing values natively → no imputation needed.**
3. **Honest validation** — 5-fold *Stratified* cross-validation with **out-of-fold** predictions and
   leak-free per-fold fitting; report macro-F1 + confusion matrix.
4. **Light tuning**, then refit on all data and write `submission.csv`.

## 0. Setup & load data

```python
import warnings; warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import lightgbm as lgb

sns.set_theme(style="whitegrid"); plt.rcParams["figure.dpi"] = 110
RANDOM_STATE = 42

train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")              # used only for the final submission
sample_submission = pd.read_csv("sample_submission.csv")

target   = "sleep_stage"
features = [c for c in train.columns if c not in ("id", target)]
X  = train[features]                          # NaNs kept on purpose — LightGBM handles them
y  = train[target].values
Xt = test[features]
classes = sorted(np.unique(y))

print("train:", train.shape, "| test:", test.shape, "| features:", len(features))
print("classes:", classes)
X.head()
```

# Exploratory Data Analysis

Only the analyses that change a modeling decision. Each cell ends with a one-line takeaway.

## 1. Target balance
Decides the baseline accuracy and whether we need stratification / class weighting.

```python
counts = pd.Series(y).value_counts().sort_index()
print(pd.DataFrame({"count": counts, "pct": (counts/len(y)*100).round(1)}))
print(f"Majority-class baseline macro-F1 ≈ {counts.max()/len(y):.3f} (we must beat this)")

sns.barplot(x=counts.index, y=counts.values, color="steelblue")
plt.title("Sleep-stage distribution"); plt.xlabel("sleep_stage"); plt.ylabel("count"); plt.show()
# Takeaway: 4 roughly balanced classes -> StratifiedKFold + macro-F1; no resampling needed.
```

## 2. Missing values
Find which columns have gaps — this decides our imputation strategy.

```python
miss = pd.DataFrame({
    "train_%": (train[features].isna().mean()*100).round(1),
    "test_%":  (test[features].isna().mean()*100).round(1),
})
print(miss[miss.sum(axis=1) > 0])
# Takeaway: only `eog_burst_index` is missing (~50% train / ~28% test).
# LightGBM routes NaNs to their own branch -> we KEEP NaNs and do NOT impute.
```

## 3. Feature signal — mutual information
Model-agnostic ranking of how much each feature tells us about the stage (captures non-linear signal).

```python
from sklearn.feature_selection import mutual_info_classif
X_mi = X.fillna(X.median())                   # MI needs no NaNs (this copy is only for the plot)
mi = pd.Series(mutual_info_classif(X_mi, y, random_state=RANDOM_STATE),
               index=features).sort_values()
mi.plot.barh(figsize=(7, 6), color="seagreen"); plt.title("Mutual information with sleep_stage")
plt.xlabel("MI"); plt.tight_layout(); plt.show()
# Takeaway: heart/EOG/respiration + several EEG bands carry most signal;
# `body_movement_index` ~ 0 (near-noise). We keep all features — trees ignore noise cheaply.
```

## 4. Class separability (PCA vs LDA)
How hard is the problem? **PCA** = unsupervised variance; **LDA** = best possible *linear* separation
(a practical ceiling for linear models). If classes overlap here, some confusion is unavoidable.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

Xs = StandardScaler().fit_transform(X.fillna(X.median()))
P = PCA(n_components=2, random_state=RANDOM_STATE).fit_transform(Xs)
L = LinearDiscriminantAnalysis(n_components=2).fit_transform(Xs, y)

fig, ax = plt.subplots(1, 2, figsize=(13, 5))
for axi, (Z, name) in zip(ax, [(P, "PCA (unsupervised)"), (L, "LDA (supervised — linear ceiling)")]):
    sc = axi.scatter(Z[:, 0], Z[:, 1], c=y, cmap="viridis", s=8, alpha=.5)
    axi.set_title(name); axi.set_xlabel("dim 1"); axi.set_ylabel("dim 2")
fig.colorbar(sc, ax=ax, label="sleep_stage", shrink=.7); plt.show()
# Takeaway: LDA shows partial, overlapping separation (esp. middle stages) ->
# expect a strong-but-not-perfect score; a non-linear model (LightGBM) should beat linear.
```

## 5. Validation & leakage sanity check
Before trusting CV we confirm: train/test ids are disjoint, there's no temporal autocorrelation
(so a shuffled split is safe), and train≈test (no drift). Adversarial AUC≈0.5 ⇒ same distribution.

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline

print("id overlap train∩test:", len(set(train.id) & set(test.id)), "(0 = disjoint, good)")
y_ord = train.sort_values("id")[target].values
chance = float((pd.Series(y).value_counts(normalize=True) ** 2).sum())
print(f"lag-1 target agreement = {(y_ord[1:]==y_ord[:-1]).mean():.3f} vs chance {chance:.3f}"
      "  -> no temporal structure, shuffled split is safe")

adv_X = pd.concat([train[features], test[features]], ignore_index=True)
adv_y = np.r_[np.zeros(len(train)), np.ones(len(test))]
adv = make_pipeline(SimpleImputer(strategy="median"),
                    RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1))
auc = cross_val_score(adv, adv_X, adv_y, cv=StratifiedKFold(5, shuffle=True, random_state=RANDOM_STATE),
                      scoring="roc_auc").mean()
print(f"adversarial AUC = {auc:.3f} -> " + ("DRIFT - investigate" if auc > 0.6 else "train ≈ test, no drift"))
```

### EDA takeaways
- **4 balanced classes** → `StratifiedKFold(5)` + **macro-F1**; majority baseline ≈ 0.27.
- **Only `eog_burst_index` missing** (~50%) → **keep NaNs**, let LightGBM handle them (no imputation).
- **Signal** spread across heart/EOG/respiration/EEG; `body_movement_index` ≈ noise (kept; harmless to trees).
- **Classes overlap** (LDA) → expect ~0.8, not 1.0; non-linear model warranted.
- **No leakage / no drift** → shuffled stratified CV is trustworthy and should transfer to the test set.

# Modeling

**Cross-validation protocol.** We score every model with **out-of-fold (OOF)** predictions on the same
`StratifiedKFold(5)`. For LightGBM each fold uses **early stopping** on its own validation part, so the
number of trees is chosen automatically and there is no leakage.

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score, accuracy_score, classification_report, ConfusionMatrixDisplay
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

def lgbm_oof(params, n_estimators=2000, stopping=50):
    # Leak-free OOF probabilities for LightGBM with per-fold early stopping.
    # Returns (oof_proba, best_iterations_per_fold).
    oof = np.zeros((len(X), len(classes))); best_iters = []
    for tr, va in cv.split(X, y):
        m = lgb.LGBMClassifier(n_estimators=n_estimators, **params)
        m.fit(X.iloc[tr], y[tr], eval_set=[(X.iloc[va], y[va])], eval_metric="multi_logloss",
              callbacks=[lgb.early_stopping(stopping, verbose=False)])
        oof[va] = m.predict_proba(X.iloc[va]); best_iters.append(m.best_iteration_)
    return oof, best_iters
```

## 6. Baseline ladder — does LightGBM earn its place?

```python
# Reference models (imputation/scaling live INSIDE the pipeline -> fit per fold, no leakage)
ref = {
    "Dummy (stratified)": DummyClassifier(strategy="stratified", random_state=RANDOM_STATE),
    "LogisticRegression": make_pipeline(SimpleImputer(strategy="median"), StandardScaler(),
                                        LogisticRegression(max_iter=2000)),
    "RandomForest":       make_pipeline(SimpleImputer(strategy="median"),
                                        RandomForestClassifier(n_estimators=400, random_state=RANDOM_STATE, n_jobs=-1)),
}
rows = []
for name, mdl in ref.items():
    pred = cross_val_predict(mdl, X, y, cv=cv, n_jobs=-1)
    rows.append({"model": name, "macro_F1": f1_score(y, pred, average="macro"),
                 "accuracy": accuracy_score(y, pred)})

LGBM_PARAMS = dict(objective="multiclass", num_class=len(classes), learning_rate=0.05,
                   num_leaves=15, min_child_samples=50, subsample=0.8, subsample_freq=1,
                   colsample_bytree=0.8, reg_lambda=2.0, class_weight="balanced",
                   random_state=RANDOM_STATE, n_jobs=-1, verbose=-1)
lgbm_oof_proba, lgbm_iters = lgbm_oof(LGBM_PARAMS)
lgbm_pred = np.array(classes)[lgbm_oof_proba.argmax(1)]
rows.append({"model": "LightGBM", "macro_F1": f1_score(y, lgbm_pred, average="macro"),
             "accuracy": accuracy_score(y, lgbm_pred)})

board = pd.DataFrame(rows).sort_values("macro_F1", ascending=False).reset_index(drop=True)
print(board.to_string(index=False))
print("\nLightGBM trees chosen per fold (early stopping):", lgbm_iters)
```

## 7. Where does LightGBM still make mistakes?

```python
print(classification_report(y, lgbm_pred, digits=3))
ConfusionMatrixDisplay.from_predictions(y, lgbm_pred, normalize="true",
                                        display_labels=classes, cmap="Blues", values_format=".2f")
plt.title("Row-normalized confusion — LightGBM (OOF)"); plt.show()
# Takeaway: class 2 is the hardest and leaks into its neighbours -> consistent with the
# overlapping LDA picture. This is genuine class overlap, not a modeling bug.
```

## 8. Light hyperparameter tuning
A small, readable comparison around sensible defaults (not an exhaustive search). We keep whatever
gives the best OOF macro-F1. Depth (`num_leaves`) is the knob that matters most here.

```python
trials = {
    "shallow (num_leaves=15)": LGBM_PARAMS,                                    # current default
    "medium  (num_leaves=31)": {**LGBM_PARAMS, "num_leaves": 31, "reg_lambda": 1.0, "min_child_samples": 30},
    "deeper  (num_leaves=63)": {**LGBM_PARAMS, "num_leaves": 63, "learning_rate": 0.03, "reg_lambda": 3.0},
    "no class_weight":         {k: v for k, v in LGBM_PARAMS.items() if k != "class_weight"},
}
results = {}
for name, p in trials.items():
    proba, _ = lgbm_oof(p)
    results[name] = f1_score(y, np.array(classes)[proba.argmax(1)], average="macro")
res = pd.Series(results).sort_values(ascending=False)
print("OOF macro-F1 by config:\n", res.round(4).to_string())

BEST_PARAMS = trials[res.index[0]]
print(f"\nChosen config: '{res.index[0]}'  ->  macro-F1 = {res.iloc[0]:.4f}")
```

## 9. Final model & submission
Refit the chosen LightGBM on **all** training data. Since there is no held-out fold at this point,
we fix the number of trees from cross-validation (mean best-iteration, +10% because the full train set
is slightly larger). Then predict the test set and write `submission.csv` in the required format.

```python
final_proba_oof, final_iters = lgbm_oof(BEST_PARAMS)
final_oof_pred = np.array(classes)[final_proba_oof.argmax(1)]
print(f"Final OOF macro-F1 = {f1_score(y, final_oof_pred, average='macro'):.4f}")

n_trees = int(np.mean(final_iters) * 1.1)
print(f"Refitting on full train with n_estimators = {n_trees}")
final_model = lgb.LGBMClassifier(n_estimators=n_trees, **BEST_PARAMS).fit(X, y)

test_pred = final_model.predict(Xt)
submission = pd.DataFrame({"id": test["id"], target: test_pred})[sample_submission.columns]
submission.to_csv("submission.csv", index=False)
print("Saved submission.csv:", submission.shape)

comp = pd.DataFrame({
    "train":    pd.Series(y).value_counts(normalize=True).sort_index(),
    "test_pred": pd.Series(test_pred).value_counts(normalize=True).sort_index(),
}).round(3)
print("\nClass distribution sanity (train vs predicted test):\n", comp)
submission.head()
```
