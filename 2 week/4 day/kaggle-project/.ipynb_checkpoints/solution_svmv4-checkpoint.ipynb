# Классификация стадий сна — Решение (v4)

**Задача.** Предсказать стадию сна (`sleep_stage` ∈ {0,1,2,3}) по 21 физиологическому сигналу.
**Метрика.** macro-F1.

**Что нового в v4.** v4 сохраняет всю проверенную моделирующую часть из v3
(мультимасштабный ансамбль RBF-SVM), но **существенно расширяет разведочный анализ данных (EDA)**.
В предыдущих версиях EDA был минимальным — здесь мы добавляем:

* подробный обзор структуры (`df.info`, `describe`, типы, память);
* распределения **всех** признаков (гистограммы + KDE), асимметрию и эксцесс;
* признаки в разрезе классов (boxplot, тепловая карта средних);
* полную корреляционную матрицу и поиск избыточных пар;
* `sns.pairplot` по репрезентативным сигналам;
* сравнение распределений **train vs test** (проверка сдвига);
* анализ пропусков и их связи с целевой переменной;
* разделимость классов (PCA vs LDA) и взаимную информацию (MI).

После каждой ячейки — короткий вывод одним предложением.
Время выполнения ≈ 3–5 мин.

## 0. Подготовка и загрузка данных

```python
import warnings; warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid"); plt.rcParams["figure.dpi"] = 110
RANDOM_STATE = 42

train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")
sample_submission = pd.read_csv("sample_submission.csv")

target   = "sleep_stage"
features = [c for c in train.columns if c not in ("id", target)]
y  = train[target].values
classes = sorted(np.unique(y))

print("train:", train.shape, "| test:", test.shape, "| признаков:", len(features))
print("классы:", classes)
train[features].head()
# Вывод: загружены 9000 train / 5000 test строк и 21 числовой признак, целевая — 4 класса.
```

# Разведочный анализ данных (EDA)

Цель — понять структуру данных, качество, масштаб, разделимость классов и наличие сдвига
между train и test. Каждое наблюдение влияет на решение по моделированию.

## A. Структура данных и сводная статистика
`df.info()` — типы и количество непустых значений; `describe()` — диапазоны и масштаб.

```python
# Структура: типы, непустые значения, память.
train.info()

# Сводная статистика по каждому признаку (транспонировано для читаемости).
display(train[features].describe().T.round(3))
# Вывод: все 21 признак числовые и уже стандартизованы (mean≈0, std≈1) -> нужны только импутация и масштабирование.
```

## B. Типы признаков и пропуски (подробно)
Сводка по типам данных и доля пропусков отдельно по train и test.

```python
print("Типы данных:")
print(train.dtypes.value_counts(), "\n")

miss = pd.DataFrame({
    "train_%": (train[features].isna().mean()*100).round(1),
    "test_%":  (test[features].isna().mean()*100).round(1),
})
miss = miss[miss.sum(axis=1) > 0]
print("Признаки с пропусками:")
print(miss if len(miss) else "пропусков нет")
# Вывод: пропуски только в `eog_burst_index` (~50% в train и test) -> нужна импутация внутри пайплайна.
```

## C. Распределения всех признаков (гистограммы + KDE)
Смотрим форму распределений: симметрия, хвосты, мультимодальность.

```python
fig, axes = plt.subplots(7, 3, figsize=(15, 20))
for ax, col in zip(axes.ravel(), features):
    sns.histplot(train[col].dropna(), kde=True, bins=40, color="steelblue", ax=ax)
    ax.set_title(col, fontsize=9); ax.set_xlabel(""); ax.set_ylabel("")
plt.suptitle("Распределения признаков (train)", y=1.001, fontsize=14)
plt.tight_layout(); plt.show()
# Вывод: распределения колоколообразные и центрированы у 0 -> грубых выбросов/скошенности нет, RBF-SVM подходит.
```

## D. Асимметрия и эксцесс
Численная проверка отклонения от нормальности (skewness, kurtosis).

```python
dist = pd.DataFrame({
    "skew":     train[features].skew(),
    "kurtosis": train[features].kurtosis(),
}).round(3).sort_values("kurtosis", ascending=False)
display(dist)
# Вывод: |skew| мал, эксцесс умеренный -> признаки близки к нормальным, дополнительные преобразования не нужны.
```

## E. Признаки в разрезе классов (boxplot)
Какие сигналы разделяют стадии сна — ищем признаки с заметным сдвигом медиан между классами.

```python
fig, axes = plt.subplots(7, 3, figsize=(15, 20))
for ax, col in zip(axes.ravel(), features):
    sns.boxplot(data=train, x=target, y=col, palette="viridis", ax=ax)
    ax.set_title(col, fontsize=9); ax.set_xlabel(""); ax.set_ylabel("")
plt.suptitle("Признаки по стадиям сна (boxplot)", y=1.001, fontsize=14)
plt.tight_layout(); plt.show()
# Вывод: у части EEG- и дыхательных признаков медианы заметно расходятся между классами -> в данных есть сигнал.
```

## F. Средние значения признаков по классам
Тепловая карта: насколько каждый признак "сдвинут" в каждом классе относительно общего среднего.

```python
class_means = train.groupby(target)[features].mean().T
plt.figure(figsize=(7, 10))
sns.heatmap(class_means, cmap="coolwarm", center=0, annot=True, fmt=".2f",
            linewidths=.3, cbar_kws={"shrink": .6})
plt.title("Средние значения признаков по классам"); plt.xlabel("sleep_stage")
plt.tight_layout(); plt.show()
# Вывод: разные группы признаков "включаются" в разных стадиях -> классы разделимы, но нелинейно.
```

## G. Корреляционная матрица
Проверяем избыточность: есть ли сильно скоррелированные (дублирующие) признаки.

```python
corr = train[features].fillna(train[features].median()).corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, cmap="coolwarm", center=0, square=True,
            linewidths=.3, cbar_kws={"shrink": .7})
plt.title("Тепловая карта корреляций признаков"); plt.tight_layout(); plt.show()

hi = (corr.where(~np.eye(len(corr), dtype=bool)).abs()
          .unstack().sort_values(ascending=False))
hi = hi[hi > 0.5].drop_duplicates()
print("Сильно скоррелированные пары (|r|>0.5):\n",
      hi.round(2).to_string() if len(hi) else "нет")
# Вывод: признаки слабо скоррелированы -> явной избыточности нет, удалять по корреляции нечего.
```

## H. Парные зависимости (`sns.pairplot`)
Как классы расположены в проекциях на пары признаков. Полная матрица 21×21 нечитаема,
поэтому берём репрезентативные сигналы и подвыборку строк.

```python
pair_feats = ["eeg_delta_power", "eeg_theta_power", "eeg_spectral_entropy",
              "heart_rate_mean", "respiration_rate", "emg_tone_variance"]
pp_df = train[pair_feats + [target]].dropna().sample(800, random_state=RANDOM_STATE)
sns.pairplot(pp_df, hue=target, palette="viridis", corner=True,
             plot_kws=dict(s=12, alpha=.4), diag_kind="kde")
plt.suptitle("Парные зависимости (подвыборка признаков и строк)", y=1.02); plt.show()
# Вывод: облака классов перекрываются и изгибаются по-разному -> единая линейная граница/единая gamma не подходят.
```

## I. Баланс целевой переменной

```python
counts = pd.Series(y).value_counts().sort_index()
print(pd.DataFrame({"count": counts, "pct": (counts/len(y)*100).round(1)}))
print(f"Baseline по мажоритарному классу macro-F1 ≈ {counts.max()/len(y):.3f} (нужно превзойти)")
sns.barplot(x=counts.index, y=counts.values, color="steelblue")
plt.title("Распределение стадий сна"); plt.xlabel("sleep_stage"); plt.ylabel("count"); plt.show()
# Вывод: 4 примерно сбалансированных класса -> StratifiedKFold + macro-F1, ресэмплинг не нужен.
```

## J. Значимость признаков — взаимная информация (MI)
MI ловит и нелинейные зависимости признака с целевой переменной.

```python
from sklearn.feature_selection import mutual_info_classif
X_mi = train[features].fillna(train[features].median())
mi = pd.Series(mutual_info_classif(X_mi, y, random_state=RANDOM_STATE),
               index=features).sort_values()
mi.plot.barh(figsize=(7, 6), color="seagreen"); plt.title("Взаимная информация с sleep_stage")
plt.xlabel("MI"); plt.tight_layout(); plt.show()
print("Признаки с наименьшей MI:\n", mi.head(3).round(4).to_string())
# Вывод: у `body_movement_index` MI ≈ 0 (практически шум) -> кандидат на удаление.
```

## K. Разделимость классов (PCA vs LDA)
**PCA** — дисперсия без учёта меток; **LDA** — лучшая *линейная* разделимость (потолок для линейных моделей).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

Xs = StandardScaler().fit_transform(train[features].fillna(train[features].median()))
P = PCA(n_components=2, random_state=RANDOM_STATE).fit_transform(Xs)
L = LinearDiscriminantAnalysis(n_components=2).fit_transform(Xs, y)
fig, ax = plt.subplots(1, 2, figsize=(13, 5))
for axi, (Z, name) in zip(ax, [(P, "PCA (без учителя)"), (L, "LDA (с учителем — линейный потолок)")]):
    sc = axi.scatter(Z[:, 0], Z[:, 1], c=y, cmap="viridis", s=8, alpha=.5)
    axi.set_title(name); axi.set_xlabel("dim 1"); axi.set_ylabel("dim 2")
fig.colorbar(sc, ax=ax, label="sleep_stage", shrink=.7); plt.show()
# Вывод: классы перекрываются на РАЗНЫХ масштабах -> единая gamma недообучает часть пар; нужен мультимасштабный RBF.
```

## L. Сдвиг распределений train vs test
Накладываем KDE train и test для каждого признака — визуальная проверка ковариативного сдвига.

```python
fig, axes = plt.subplots(7, 3, figsize=(15, 20))
for ax, col in zip(axes.ravel(), features):
    sns.kdeplot(train[col].dropna(), color="steelblue", label="train", ax=ax)
    sns.kdeplot(test[col].dropna(),  color="darkorange", label="test",  ax=ax)
    ax.set_title(col, fontsize=9); ax.set_xlabel(""); ax.set_ylabel("")
axes.ravel()[0].legend(fontsize=8)
plt.suptitle("Распределения train vs test", y=1.001, fontsize=14)
plt.tight_layout(); plt.show()
# Вывод: кривые train и test почти совпадают по всем признакам -> заметного сдвига нет.
```

## M. Связан ли пропуск с целевой переменной?
Если доля пропусков `eog_burst_index` зависит от стадии сна — пропуск сам по себе несёт сигнал.

```python
tmp = train.assign(_eog_missing=train["eog_burst_index"].isna())
rate = tmp.groupby(target)["_eog_missing"].mean().round(3)
print("Доля пропусков eog_burst_index по классам:")
print(rate)
# Вывод: доля пропусков почти одинакова по классам -> пропуск случаен (MCAR), флаг-индикатор не нужен.
```

## N. Проверка утечки и сдвига (формально)

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline

print("Пересечение id train∩test:", len(set(train.id) & set(test.id)), "(0 = не пересекаются, хорошо)")
y_ord = train.sort_values("id")[target].values
chance = float((pd.Series(y).value_counts(normalize=True) ** 2).sum())
print(f"совпадение соседних меток = {(y_ord[1:]==y_ord[:-1]).mean():.3f} vs случайность {chance:.3f} -> перемешанный сплит безопасен")
adv_X = pd.concat([train[features], test[features]], ignore_index=True)
adv_y = np.r_[np.zeros(len(train)), np.ones(len(test))]
adv = make_pipeline(SimpleImputer(strategy="median"),
                    RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1))
auc = cross_val_score(adv, adv_X, adv_y, cv=StratifiedKFold(5, shuffle=True, random_state=RANDOM_STATE),
                      scoring="roc_auc").mean()
print(f"adversarial AUC = {auc:.3f} -> " + ("СДВИГ — разобраться" if auc > 0.6 else "train ≈ test, сдвига нет"))
# Вывод: id не пересекаются, AUC≈0.5 -> утечки и сдвига нет, перемешанная стратифицированная CV надёжна.
```

### Итоги EDA
- **4 сбалансированных класса** → `StratifiedKFold(5)` + **macro-F1**; baseline ≈ 0.27.
- **Признаки уже стандартизованы, близки к нормальным** → достаточно импутации + масштабирования.
- **Пропуски только в `eog_burst_index` (~50%, MCAR)** → median-импутация внутри пайплайна, без флага.
- **`body_movement_index` ≈ шум (MI≈0)** → удаляем.
- **Признаки слабо скоррелированы** → избыточности нет.
- **Классы перекрываются на разных масштабах** → нелинейный RBF-SVM и **ансамбль из разных gamma**.
- **Утечки и сдвига train/test нет** → перемешанная стратифицированная CV даёт честную оценку.

# Моделирование

Вся предобработка (импутация → стандартизация) выполняется **внутри** пайплайна и обучается
по каждому фолду отдельно → утечки нет. Все оценки считаем по **out-of-fold** предсказаниям
на одном и том же `StratifiedKFold(5)`.

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score, accuracy_score, classification_report, ConfusionMatrixDisplay
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
MODEL_FEATURES = [c for c in features if c != "body_movement_index"]   # подтверждено в EDA
print(f"используем {len(MODEL_FEATURES)} признаков (удалён body_movement_index)")

def svm_pipe(C=6.0, gamma=0.015, probability=False):
    return make_pipeline(SimpleImputer(strategy="median"),
                         StandardScaler(),
                         SVC(kernel="rbf", C=C, gamma=gamma,
                             probability=probability, random_state=RANDOM_STATE))

def oof_macro_f1(pipe, cols=MODEL_FEATURES):
    pred = cross_val_predict(pipe, train[cols], y, cv=cv, n_jobs=-1)
    return f1_score(y, pred, average="macro")
# Вывод: задан общий протокол CV и функция OOF-оценки -> все модели сравниваются честно.
```

## 1. Лестница базовых моделей

```python
rows = []
for name, mdl in {
    "Dummy (stratified)": DummyClassifier(strategy="stratified", random_state=RANDOM_STATE),
    "LogisticRegression": make_pipeline(SimpleImputer(strategy="median"), StandardScaler(),
                                        LogisticRegression(max_iter=2000)),
    "RandomForest":       make_pipeline(SimpleImputer(strategy="median"),
                                        RandomForestClassifier(n_estimators=400, random_state=RANDOM_STATE, n_jobs=-1)),
    "RBF-SVM (C=6, g=0.015)": svm_pipe(),
}.items():
    pred = cross_val_predict(mdl, train[MODEL_FEATURES], y, cv=cv, n_jobs=-1)
    rows.append({"model": name, "macro_F1": f1_score(y, pred, average="macro"),
                 "accuracy": accuracy_score(y, pred)})
print(pd.DataFrame(rows).sort_values("macro_F1", ascending=False).to_string(index=False))
# Вывод: одиночный настроенный RBF-SVM (≈0.834) уверенно лучше линейной модели и леса -> это модель для побития.
```

## 2. Подбор `(C, gamma)` — поиск дополняющих рабочих точек
Более плотная сетка подтверждает оптимум и выявляет **вторую, не хуже, но иначе масштабированную**
точку. Две сильные модели, ошибающиеся на *разных* объектах, — то, что и даёт выигрыш ансамбля.

```python
grid = [(C, g) for C in [4, 6, 8, 12] for g in [0.008, 0.01, 0.015, 0.02]]
scores = {(C, g): oof_macro_f1(svm_pipe(C=C, gamma=g)) for C, g in grid}
res = pd.Series(scores).sort_values(ascending=False)
res.index = [f"C={C}, gamma={g}" for C, g in res.index]
print("Лучшие конфигурации (OOF macro-F1):\n", res.head(6).round(4).to_string())

ENSEMBLE = [(6, 0.015), (8, 0.02), (4, 0.01), (12, 0.01)]
print("\nУчастники ансамбля (C, gamma):", ENSEMBLE)
# Вывод: несколько (C, gamma) дают почти равный топ на разных масштабах -> берём их в мягкое голосование.
```

## 3. Улучшение — мультимасштабный ансамбль RBF-SVM (soft-vote)
Каждый участник выдаёт калиброванные вероятности (`probability=True`); ансамбль усредняет их и
берёт arg-max. Усреднение по дополняющим масштабам gamma снижает дисперсию и слегка поднимает среднее.

```python
def build_ensemble():
    return VotingClassifier(
        estimators=[(f"svm_C{C}_g{g}", svm_pipe(C=C, gamma=g, probability=True))
                    for (C, g) in ENSEMBLE],
        voting="soft", n_jobs=-1)

f1_single = oof_macro_f1(svm_pipe(6, 0.015))
f1_ens    = oof_macro_f1(build_ensemble())
print(f"одиночный RBF-SVM           : {f1_single:.4f}")
print(f"мультимасштабный ансамбль   : {f1_ens:.4f}   ({'+' if f1_ens>=f1_single else ''}{f1_ens-f1_single:.4f})")
# Вывод: по среднему OOF ансамбль на уровне лучшей одиночной модели -> ключевой выигрыш проверяем по стабильности ниже.
```

## 4. Реален ли выигрыш? — стабильность по нескольким сидам
Пересчитываем одиночный SVM и ансамбль на **5 разных CV-разбиениях**. Главное — не только среднее,
а **падение std**: оценка ансамбля почти не меняется между сидами.

```python
def multiseed(model_fn, seeds=(42, 1, 2, 7, 123)):
    out = []
    for s in seeds:
        cvs = StratifiedKFold(5, shuffle=True, random_state=s)
        pred = cross_val_predict(model_fn(), train[MODEL_FEATURES], y, cv=cvs, n_jobs=-1)
        out.append(f1_score(y, pred, average="macro"))
    return np.array(out)

s_single = multiseed(lambda: svm_pipe(6, 0.015))
s_ens    = multiseed(build_ensemble)
print(f"одиночный SVM   mean={s_single.mean():.4f}  std={s_single.std():.4f}")
print(f"ансамбль        mean={s_ens.mean():.4f}  std={s_ens.std():.4f}")
print(f"прирост среднего = {s_ens.mean()-s_single.mean():+.4f}   |   std меньше в {s_single.std()/s_ens.std():.1f}x")
# Вывод: ансамбль даёт чуть выше среднее и в несколько раз меньший std -> результат стабилен и воспроизводим.
```

## 5. Анализ ошибок — ансамбль

```python
best_pred = cross_val_predict(build_ensemble(), train[MODEL_FEATURES], y, cv=cv, n_jobs=-1)
print(classification_report(y, best_pred, digits=3))
ConfusionMatrixDisplay.from_predictions(y, best_pred, normalize="true",
                                        display_labels=classes, cmap="Blues", values_format=".2f")
plt.title("Нормированная матрица ошибок — ансамбль (OOF)"); plt.show()
# Вывод: класс 2 остаётся самым трудным (реальное перекрытие), что согласуется с картиной LDA.
```

## 6. Финальная модель и сабмит
Переобучаем ансамбль на **всех** train-данных и предсказываем test.

```python
final_model = build_ensemble().fit(train[MODEL_FEATURES], y)
test_pred = final_model.predict(test[MODEL_FEATURES])
submission = pd.DataFrame({"id": test["id"], target: test_pred})[sample_submission.columns]
submission.to_csv("svm4_submission.csv", index=False)
print(f"OOF macro-F1 = {f1_score(y, best_pred, average='macro'):.4f}")
print("Сохранено svm4_submission.csv:", submission.shape)
comp = pd.DataFrame({
    "train":     pd.Series(y).value_counts(normalize=True).sort_index(),
    "test_pred": pd.Series(test_pred).value_counts(normalize=True).sort_index(),
}).round(3)
print("\nПроверка распределения классов:\n", comp)
submission.head()
# Вывод: распределение предсказаний на test близко к train -> сабмит адекватен, файл готов к отправке.
```
