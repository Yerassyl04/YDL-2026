# Sleep-stage classification — Stacking ensemble

Third companion notebook to `inception_random forest.ipynb` (a tuned **RandomForest**,
OOF macro-F1 ≈ 0.80) and `boosting.ipynb` (a soft-voting **booster blend**, OOF macro-F1 ≈ 0.826).
Same data, same leak-free validation, same competition metric (**macro-F1**, 4 balanced classes).

The modelling idea here is **stacked generalization**: instead of *averaging* model
probabilities (the soft-voting blend of `boosting.ipynb`), we train a **meta-learner** that
learns *how* to combine the base models' predictions — it can up-weight whichever base model
is most trustworthy in each region of probability space.

**Pipeline**
1. preprocess (identical recipe to the other two notebooks)
2. a **diverse** zoo of base learners (trees + boosters + linear + distance-based)
3. score each base learner out-of-fold (the ladder)
4. generate **out-of-fold probability meta-features** (leak-free — the heart of stacking)
5. select & train a **meta-learner** on those meta-features
6. cross-check against sklearn's `StackingClassifier`
7. per-class threshold optimization for macro-F1
8. stability check → final refit → `submission_stacking.csv`

> **Why stacking can beat blending:** a fixed average treats every base model equally
> everywhere. A meta-learner can discover, e.g., "trust CatBoost for class 2 but the linear
> model near the 0/1 boundary." It only helps when base learners make *diverse* errors, so we
> deliberately include models from different families.

## 0. Dependencies

The booster libraries (`xgboost`, `lightgbm`, `catboost`) are optional base learners — the
notebook filters to whatever is importable, exactly like `boosting.ipynb`.

```python
# Run once if needed (uncomment):
# %pip install -q xgboost lightgbm catboost

import importlib
_avail = {}
for _lib in ("xgboost", "lightgbm", "catboost"):
    try:
        importlib.import_module(_lib)
        _avail[_lib] = True
    except ImportError:
        _avail[_lib] = False
print("available boosting libs:", _avail)
print("(RandomForest / ExtraTrees / HistGB / LogReg / KNN come from sklearn, always available)")
```

## 1. Load data

Identical setup to the other notebooks: 21 float features, integer target `sleep_stage`
(0–3), `test.csv` used only for the final submission.

```python
import numpy as np
import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
sample_submission = pd.read_csv("sample_submission.csv")

target = "sleep_stage"
num_features = [c for c in train.columns if c not in ("id", target)]
classes = sorted(train[target].unique())

print("train:", train.shape, "| test:", test.shape)
print("features:", len(num_features), "| classes:", classes)
print("class balance:\n", train[target].value_counts(normalize=True).sort_index().round(3))
```

## 2. Preprocessing & validation setup

The EDA in `inception_random forest.ipynb` established everything we need: only
`eog_burst_index` has missing values (~50% train / ~28% test); classes are balanced;
features are pre-standardized; train≈test (adversarial AUC≈0.5, no drift); ids are disjoint
with no temporal autocorrelation → a shuffled `StratifiedKFold` is the correct CV.

We keep **raw features with NaNs intact** as the canonical matrix. Each base learner then
handles missingness the way it prefers:
- native-NaN models (HistGB, XGBoost, LightGBM, CatBoost) consume NaNs directly;
- models that can't (RandomForest, ExtraTrees, LogReg, KNN) get a `SimpleImputer` **inside
  their own pipeline**, so imputation is fit per-fold and never leaks.

Using the **same** `cv` object for every out-of-fold computation is what makes the meta-features
aligned and the stack leak-free.

```python
from sklearn.model_selection import StratifiedKFold

# Canonical raw matrix: keep NaNs so native-NaN learners can exploit them.
Xr = train[num_features].copy()
yr = train[target].values
labels = np.array(sorted(np.unique(yr)))

# Leak-free CV: shuffled stratified 5-fold (justified by inception §13). Shared everywhere.
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("NaNs kept in Xr (only eog_burst_index):", int(Xr.isna().sum().sum()))
print("Xr:", Xr.shape, "| classes:", labels.tolist())
```

## 3. Feature engineering (shared `build()` contract)

Same contract as the other two notebooks: one function with **identical logic for train and
test**, so the winning recipe is reused verbatim at submission time. The companion notebooks
found the plain feature set competitive (RF liked a missing-flag; boosters liked base-21), so
we expose the same knobs and let a quick probe pick the recipe for the stack.

```python
import itertools
from sklearn.base import clone
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import HistGradientBoostingClassifier

TOP = ["heart_rate_variability", "eog_burst_index", "eog_movement_density",
       "eeg_alpha_power", "respiration_rate"]

def build(df, drop_bmi=False, interactions=False, flag=False):
    """Build a feature matrix; identical logic for train and test (NaNs preserved)."""
    cols = [c for c in num_features if not (drop_bmi and c == "body_movement_index")]
    out = df[cols].copy()
    if interactions:
        for a, b in itertools.combinations(TOP, 2):
            out[f"{a}*{b}"] = df[a] * df[b]
    if flag:
        out["eog_burst_index_missing"] = df["eog_burst_index"].isna().astype(int)
    return out

# Fast probe to pick a recipe (HistGB handles NaN natively).
probe = HistGradientBoostingClassifier(max_iter=300, learning_rate=0.06,
                                       max_leaf_nodes=31, random_state=42)
configs = {
    "base (all 21)":   dict(),
    "+ flag":          dict(flag=True),
    "drop bmi + flag": dict(drop_bmi=True, flag=True),
}
fe = pd.Series({
    name: cross_val_score(clone(probe), build(train, **kw), yr,
                          cv=cv, scoring="f1_macro", n_jobs=-1).mean()
    for name, kw in configs.items()
}).sort_values(ascending=False)
print("CV macro-F1 by feature set (HistGB probe):")
print(fe.round(4).to_string())

BEST_KW = configs[fe.index[0]]
X_fe = build(train, **BEST_KW)
print(f"\nChosen feature set: '{fe.index[0]}'  ->  {BEST_KW}")
print("n features:", X_fe.shape[1])
```

## 4. Base-learner zoo

Stacking lives or dies on **error diversity**, so the zoo deliberately spans four model
families:

| family | model | why it's here |
|---|---|---|
| bagged trees | `RandomForest`, `ExtraTrees` | strong, low-variance, decorrelated splits |
| boosted trees | `HistGB`, `XGBoost`, `LightGBM`, `CatBoost` | the dataset's best single models |
| linear | `LogisticRegression` | captures the linear/ordinal signal (inception §11), very different bias |
| distance | `KNN` | local structure the trees miss |

Native-NaN boosters take raw NaNs; the rest impute inside their own pipeline (leak-free per
fold). Each is a single estimator/pipeline that exposes `predict_proba` — all stacking needs.

```python
from sklearn.ensemble import (RandomForestClassifier, ExtraTreesClassifier,
                              HistGradientBoostingClassifier)
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

RNG = 42

def make_base_learners():
    """Fresh, diverse base estimators. Native-NaN models take raw NaNs; the rest
    impute (and scale, where it matters) inside their own leak-free pipeline."""
    m = {}

    # --- bagged trees (need imputation: sklearn RF/ET reject NaN) ---
    m["RandomForest"] = make_pipeline(
        SimpleImputer(strategy="median"),
        RandomForestClassifier(n_estimators=400, max_features=0.3,
                               class_weight="balanced", random_state=RNG, n_jobs=-1),
    )
    m["ExtraTrees"] = make_pipeline(
        SimpleImputer(strategy="median"),
        ExtraTreesClassifier(n_estimators=500, max_features=0.3,
                             class_weight="balanced", random_state=RNG, n_jobs=-1),
    )

    # --- boosted trees (native NaN) ---
    m["HistGB"] = HistGradientBoostingClassifier(
        max_iter=400, learning_rate=0.05, max_leaf_nodes=31,
        l2_regularization=1.0, random_state=RNG)

    if _avail.get("xgboost"):
        from xgboost import XGBClassifier
        m["XGBoost"] = XGBClassifier(
            n_estimators=400, learning_rate=0.05, max_depth=6,
            subsample=0.8, colsample_bytree=0.8, reg_lambda=1.0,
            objective="multi:softprob", num_class=len(classes),
            tree_method="hist", eval_metric="mlogloss", random_state=RNG, n_jobs=-1)

    if _avail.get("lightgbm"):
        from lightgbm import LGBMClassifier
        m["LightGBM"] = LGBMClassifier(
            n_estimators=400, learning_rate=0.05, num_leaves=31,
            subsample=0.8, colsample_bytree=0.8, reg_lambda=1.0,
            objective="multiclass", num_class=len(classes),
            random_state=RNG, n_jobs=-1, verbose=-1)

    if _avail.get("catboost"):
        from catboost import CatBoostClassifier
        m["CatBoost"] = CatBoostClassifier(
            iterations=400, learning_rate=0.05, depth=6, l2_leaf_reg=3.0,
            loss_function="MultiClass", random_seed=RNG, verbose=0)

    # --- linear: very different bias; impute + scale for LogReg ---
    m["LogReg"] = make_pipeline(
        SimpleImputer(strategy="median"), StandardScaler(),
        LogisticRegression(C=1.0, max_iter=2000, class_weight="balanced"))

    # --- distance-based: local structure; impute + scale for KNN ---
    m["KNN"] = make_pipeline(
        SimpleImputer(strategy="median"), StandardScaler(),
        KNeighborsClassifier(n_neighbors=30, weights="distance"))

    return m

BASE_LEARNERS = make_base_learners()
print("base learners:", list(BASE_LEARNERS))
```

## 5. Base-learner ladder (macro-F1, out-of-fold)

Score each base learner with **out-of-fold** predictions on the shared `cv`. This both ranks
the members and, crucially, reuses the *same* OOF probabilities as the stacking meta-features
in §6 — so the ladder is computed once and feeds everything downstream.

We report macro-F1 + accuracy, and a **mean pairwise correlation of OOF errors** per model:
lower error-correlation ⇒ more complementary ⇒ more value to the stack.

```python
import time
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score, accuracy_score, classification_report

# Cache OOF probabilities once; reuse as meta-features in §6.
oof_proba = {}
rows = []
for name, mdl in BASE_LEARNERS.items():
    t0 = time.perf_counter()
    p = cross_val_predict(clone(mdl), X_fe, yr, cv=cv, method="predict_proba", n_jobs=-1)
    dt = time.perf_counter() - t0
    oof_proba[name] = p
    pred = labels[p.argmax(1)]
    rows.append({"model": name,
                 "macro_F1": f1_score(yr, pred, average="macro"),
                 "accuracy": accuracy_score(yr, pred),
                 "fit_sec": round(dt, 1)})

ladder = pd.DataFrame(rows).sort_values("macro_F1", ascending=False).reset_index(drop=True)
print(ladder.to_string(index=False))

# Error-diversity: correlation of (pred != truth) indicators across models.
err = pd.DataFrame({n: (labels[p.argmax(1)] != yr).astype(int) for n, p in oof_proba.items()})
err_corr = err.corr()
mean_off_diag = (err_corr.sum(1) - 1) / (len(err_corr) - 1)
print("\nmean pairwise error-correlation (lower = more complementary):")
print(mean_off_diag.sort_values().round(3).to_string())
```

## 6. Build the out-of-fold meta-features

This is the core of stacking. For each base learner we already have its **out-of-fold**
class-probability matrix (n×4) from §5 — predictions for each row made by a model that
*never saw that row in training*. Concatenating them horizontally gives the meta-feature
matrix `Z_oof` (n × 4·K), a clean, leak-free representation of "what every base model thinks."

The meta-learner is trained on `Z_oof`. Because the columns are OOF predictions, the
meta-learner cannot cheat by reading a base model that memorised the row.

We keep **all four** probability columns per learner (they sum to 1, so one is technically
redundant); the regularized meta-learner handles the collinearity, and keeping them is the
standard, simplest choice.

```python
def stack_matrix(proba_dict, members):
    """Horizontally stack per-model probability matrices into meta-features."""
    return np.hstack([proba_dict[n] for n in members])

members = list(oof_proba)                      # all base learners participate
Z_oof = stack_matrix(oof_proba, members)
meta_cols = [f"{n}_p{c}" for n in members for c in labels]

print("meta-feature matrix Z_oof:", Z_oof.shape, f"({len(members)} models x {len(labels)} classes)")
print("first few meta-columns:", meta_cols[:6], "...")
```

## 7. Select & train the meta-learner

The meta-learner should be **simple and well-regularized** — its job is to weight base
predictions, not to re-learn the problem (a complex meta-learner overfits the base models'
quirks). We compare a few candidates with **out-of-fold predictions on `Z_oof`** using the
same `cv`, scored by macro-F1:

- `LogReg` (L2) — the classic, robust stacker.
- `LogReg` (lower C) — stronger regularization.
- `RidgeClassifier`-style via multinomial LogReg with small C.
- a shallow `HistGB` meta-learner — can capture non-linear "trust regions," at overfit risk.

> **Note on honesty:** `Z_oof` is leak-free w.r.t. the *base* learners, so this is the
> standard stacking estimate. It can be mildly optimistic because the meta-learner is
> evaluated on the same OOF rows it trains on across folds; §8 cross-checks the whole stack
> with sklearn's nested `StackingClassifier`, and §10 confirms stability.

```python
from sklearn.linear_model import LogisticRegression

meta_candidates = {
    "LogReg (C=1.0)":  LogisticRegression(C=1.0, max_iter=2000, class_weight="balanced"),
    "LogReg (C=0.3)":  LogisticRegression(C=0.3, max_iter=2000, class_weight="balanced"),
    "LogReg (C=0.1)":  LogisticRegression(C=0.1, max_iter=2000, class_weight="balanced"),
    "HistGB (shallow)": HistGradientBoostingClassifier(
        max_iter=200, learning_rate=0.05, max_leaf_nodes=15,
        l2_regularization=2.0, random_state=RNG),
}

meta_rows = []
for name, meta in meta_candidates.items():
    pred = cross_val_predict(clone(meta), Z_oof, yr, cv=cv, n_jobs=-1)
    meta_rows.append({"meta_learner": name,
                      "macro_F1": f1_score(yr, pred, average="macro"),
                      "accuracy": accuracy_score(yr, pred)})

meta_board = pd.DataFrame(meta_rows).sort_values("macro_F1", ascending=False).reset_index(drop=True)
print(meta_board.to_string(index=False))

BEST_META_NAME = meta_board.iloc[0]["meta_learner"]
meta_learner = clone(meta_candidates[BEST_META_NAME])
print(f"\nChosen meta-learner: {BEST_META_NAME}")

# OOF probabilities of the full stack (meta-learner over Z_oof) — feeds thresholding in §9.
stack_oof_proba = cross_val_predict(clone(meta_learner), Z_oof, yr, cv=cv,
                                    method="predict_proba", n_jobs=-1)
stack_oof_pred = labels[stack_oof_proba.argmax(1)]
f1_stack = f1_score(yr, stack_oof_pred, average="macro")
print(f"\nStack OOF macro-F1 (argmax) = {f1_stack:.4f}")
print("\nOOF classification report — stack:\n")
print(classification_report(yr, stack_oof_pred, digits=3))
```

## 8. Cross-check with sklearn `StackingClassifier`

To make sure the manual stack isn't accidentally leaking, we rebuild the same idea with
sklearn's `StackingClassifier`, which does the OOF construction internally (`cv=cv`) and
refits base learners on full data for inference. Wrapping it in `cross_val_predict` gives a
**fully nested**, unbiased estimate of the stack.

This is expensive (nested CV over many base learners), so we keep it as a one-shot
validation. If the manual-stack OOF (§7) and the nested `StackingClassifier` agree closely,
the manual meta-features are trustworthy. We also try `passthrough=True`, which feeds the
original features to the meta-learner alongside base predictions.

```python
from sklearn.ensemble import StackingClassifier

# Use a lighter base set for the nested cross-check to keep runtime sane.
stack_est = StackingClassifier(
    estimators=[(n, clone(m)) for n, m in BASE_LEARNERS.items()],
    final_estimator=clone(meta_learner),
    stack_method="predict_proba", cv=cv, n_jobs=-1, passthrough=False,
)

# Nested OOF estimate of the whole stack (honest but slow).
nested_pred = cross_val_predict(stack_est, X_fe, yr, cv=cv, n_jobs=1)
f1_nested = f1_score(yr, nested_pred, average="macro")
print(f"Manual-stack OOF macro-F1   = {f1_stack:.4f}")
print(f"Nested StackingClassifier   = {f1_nested:.4f}")
print(f"gap (manual - nested)       = {f1_stack - f1_nested:+.4f}",
      "  (small gap => manual meta-features are trustworthy)")
```

## 9. Per-class threshold optimization (macro-F1)

`argmax(proba)` maximizes accuracy, not macro-F1. As in the companion notebooks, we learn a
per-class probability **multiplier** by coordinate ascent on the stack's **out-of-fold**
probabilities (leak-free), then predict with `argmax(proba * class_mult)`. The learned
`class_mult` is reused for the test predictions in §11.

```python
def fit_class_mult(proba, y, labels, grid=None):
    """Coordinate-ascent search for per-class probability multipliers that maximize macro-F1."""
    if grid is None:
        grid = np.round(np.linspace(0.5, 2.0, 31), 3)
    mult = np.ones(proba.shape[1])
    improved = True
    while improved:
        improved = False
        for k in range(len(mult)):
            best_m = mult[k]
            best_s = f1_score(y, labels[(proba * mult).argmax(1)], average="macro")
            for m in grid:
                cand = mult.copy(); cand[k] = m
                s = f1_score(y, labels[(proba * cand).argmax(1)], average="macro")
                if s > best_s:
                    best_s, best_m = s, m
            if best_m != mult[k]:
                mult[k] = best_m; improved = True
    return mult

class_mult = fit_class_mult(stack_oof_proba, yr, labels)
opt_pred = labels[(stack_oof_proba * class_mult).argmax(1)]

f1_argmax = f1_score(yr, stack_oof_pred, average="macro")
f1_opt = f1_score(yr, opt_pred, average="macro")
print(f"Stack OOF macro-F1  argmax    = {f1_argmax:.4f}")
print(f"Stack OOF macro-F1  optimized = {f1_opt:.4f}   (+{f1_opt - f1_argmax:.4f})")
print("class multipliers:", dict(zip(labels.tolist(), class_mult.round(3).tolist())))
print("\nOptimized-threshold OOF report:\n")
print(classification_report(yr, opt_pred, digits=3))
```

## 10. Stability check (seed variation)

The stack inherits stochasticity from its tree/booster bases. We rebuild the **whole stack**
(re-seed every base learner that has a seed, recompute OOF meta-features, refit the
meta-learner) across a few seeds and report OOF macro-F1 as mean ± std — confirming the gain
is real, not a lucky draw.

```python
def set_seed(est, s):
    """Re-seed an estimator/pipeline wherever a seed param exists."""
    e = clone(est)
    params = e.get_params()
    for p in list(params):
        if p.split("__")[-1] in ("random_state", "random_seed", "seed"):
            try:
                e.set_params(**{p: s})
            except ValueError:
                pass
    return e

seeds = [0, 1, 2]
seed_scores = []
for s in seeds:
    # Re-seed bases, regenerate OOF meta-features, refit meta-learner OOF.
    z_parts = []
    for name, mdl in BASE_LEARNERS.items():
        p = cross_val_predict(set_seed(mdl, s), X_fe, yr, cv=cv,
                              method="predict_proba", n_jobs=-1)
        z_parts.append(p)
    Z_s = np.hstack(z_parts)
    pred = cross_val_predict(set_seed(meta_learner, s), Z_s, yr, cv=cv, n_jobs=-1)
    seed_scores.append(f1_score(yr, pred, average="macro"))

seed_scores = np.array(seed_scores)
print("per-seed stack OOF macro-F1:", dict(zip(seeds, seed_scores.round(4).tolist())))
print(f"mean = {seed_scores.mean():.4f}  std = {seed_scores.std():.4f}")
```

## 11. Final model & submission

Refit the full stack on **all** of train and predict `test`:

1. refit every base learner on full train, predict **test** probabilities;
2. refit each base learner's **OOF** probabilities were already computed on train → that is
   the meta-learner's training set `Z_oof`;
3. fit the meta-learner on `Z_oof` (train) and apply it to the stacked **test** probabilities;
4. apply the §9 per-class multipliers (`argmax(proba * class_mult)`), the rule validated OOF;
5. write `submission_stacking.csv` in `sample_submission` format.

This is the canonical, fully leak-free stacking refit: base learners are trained on all of
train for *inference*, but the meta-learner only ever learned from *out-of-fold* base
predictions.

```python
# 1) Build train/test feature matrices with the SAME recipe.
X_full   = build(train, **BEST_KW)
X_test_f = build(test,  **BEST_KW)
assert list(X_full.columns) == list(X_test_f.columns), "train/test feature mismatch!"

# 2) Refit each base learner on full train, predict test probabilities.
test_proba_by_model = {}
for name, mdl in BASE_LEARNERS.items():
    fitted = clone(mdl).fit(X_full, yr)
    test_proba_by_model[name] = fitted.predict_proba(X_test_f)

# 3) Assemble test meta-features in the SAME member order as Z_oof.
Z_test = stack_matrix(test_proba_by_model, members)
assert Z_test.shape[1] == Z_oof.shape[1], "meta-feature width mismatch!"

# 4) Fit meta-learner on OOF meta-features (train), apply to test meta-features.
final_meta = clone(meta_learner).fit(Z_oof, yr)
test_proba = final_meta.predict_proba(Z_test)
test_pred = labels[(test_proba * class_mult).argmax(1)]

# 5) Write submission.
submission = pd.DataFrame({"id": test["id"], target: test_pred})[sample_submission.columns]
submission.to_csv("submission_stacking.csv", index=False)

print(f"Approach: stack[{'+'.join(members)}] -> {BEST_META_NAME}")
print(f"Stack OOF macro-F1 (with thresholds) = {f1_opt:.4f}")
print("Saved submission_stacking.csv:", submission.shape)
print(submission.head(), "\n")

comp = pd.DataFrame({
    "train":            train[target].value_counts(normalize=True).sort_index(),
    "test_argmax":      pd.Series(labels[test_proba.argmax(1)]).value_counts(normalize=True).sort_index(),
    "test_thresholded": submission[target].value_counts(normalize=True).sort_index(),
}).round(3)
print("Class distribution (train vs predicted test):")
print(comp)
```

## Takeaways

- **Stacking vs blending.** The soft-voting blend in `boosting.ipynb` averages base
  probabilities with fixed weights; here a **meta-learner** *learns* the combination from
  out-of-fold predictions, so it can weight base models differently per class/region.
- **Diversity is the fuel.** The §5 error-correlation table shows which base learners are
  complementary — the linear (`LogReg`) and distance (`KNN`) members add bias diversity the
  trees lack, even though they're individually weaker.
- **Leak-free by construction.** Meta-features are built from *out-of-fold* base predictions
  (§6), the meta-learner trains only on those, and the §8 nested `StackingClassifier`
  cross-check confirms the manual estimate isn't optimistic.
- **Same metric discipline** as the companions: per-class threshold optimization (§9) targets
  macro-F1 directly, and the seed sweep (§10) confirms stability.
- **Compare the three notebooks:** RandomForest ≈ 0.80 → booster blend ≈ 0.826 → this stack.
  Stacking pays off when the base learners are both *strong* and *diverse*; if the stack only
  ties the boosting blend, the boosters were already capturing most of the signal.
