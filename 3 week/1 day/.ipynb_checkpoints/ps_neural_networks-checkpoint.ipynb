## Part I - Data

### Question 1: Load data + exploratory analysis

We'll use the [CIFAR-10](https://en.wikipedia.org/wiki/CIFAR-10) dataset, a classic
benchmark in ML. It has 10 classes; we'll use just two of them for a binary task.

Helper code to load the data is provided below.

**Your tasks:**
- Create a subset of CIFAR-10 keeping only the **automobile** and **truck** classes.
- Select 9 random images from your training subset and plot them in a 3 x 3 grid,
  each titled with its label.

```python
from torchvision import datasets, transforms
from torch.utils.data import Subset, DataLoader, TensorDataset
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(42)
np.random.seed(42)

def loadCifar10(data_path):
    "load CIFAR-10 train and test datasets (pixels scaled to [0, 1])"
    transform = transforms.Compose([transforms.ToTensor()])
    cifar_train = datasets.CIFAR10(data_path, train=True, download=True, transform=transform)
    cifar_test = datasets.CIFAR10(data_path, train=False, download=True, transform=transform)
    return cifar_train, cifar_test
```

```python
cifar_train, cifar_test = loadCifar10("./data/")
print(cifar_train.classes)
```

```python
### YOUR CODE HERE
# Classes we want
selected_classes = [1, 9]   # automobile, truck

# Find indices of automobile and truck images
train_indices = [i for i, (_, label) in enumerate(cifar_train)
                 if label in selected_classes]

test_indices = [i for i, (_, label) in enumerate(cifar_test)
                if label in selected_classes]

# Create subsets
train_subset = Subset(cifar_train, train_indices)
test_subset = Subset(cifar_test, test_indices)

# Label mapping
label_names = {1: "Automobile", 9: "Truck"}

# Plot 9 random images
random_indices = np.random.choice(len(train_subset), 9, replace=False)

plt.figure(figsize=(8, 8))

for i, idx in enumerate(random_indices):
    image, label = train_subset[idx]

    plt.subplot(3, 3, i + 1)
    plt.imshow(image.permute(1, 2, 0))
    plt.title(label_names[label])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

### Question 2: Prepare the data for a feedforward network

A feedforward network takes a flat **vector**, not a 3 x 32 x 32 image. Get the data
ready:

- Build flattened feature tensors `x_train`, `x_test` (each image becomes a length-3072
  vector) and integer label tensors `y_train`, `y_test` (0 = automobile, 1 = truck).
- Wrap them in `DataLoader`s with batch size 32 (shuffle the training loader).

**Short answer:** how many input features does each image become, and why must we
flatten the image before feeding it to a fully-connected network?

```python
### YOUR CODE HERE
# Convert subset into tensors
x_train = []
y_train = []

for image, label in train_subset:
    # Flatten image (3 x 32 x 32 -> 3072)
    x_train.append(image.view(-1))

    # Convert labels:
    # automobile -> 0
    # truck -> 1
    y_train.append(0 if label == 1 else 1)

x_train = torch.stack(x_train)
y_train = torch.tensor(y_train, dtype=torch.long)


x_test = []
y_test = []

for image, label in test_subset:
    x_test.append(image.view(-1))
    y_test.append(0 if label == 1 else 1)

x_test = torch.stack(x_test)
y_test = torch.tensor(y_test, dtype=torch.long)


print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)


# Create TensorDatasets
train_dataset = TensorDataset(x_train, y_train)
test_dataset = TensorDataset(x_test, y_test)

# DataLoaders
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False
)
```

*Your short answer here.*

Each image becomes 3072 input features because a CIFAR-10 image has 3 color channels (RGB) and each channel is 32 × 32 pixels, so:

3×32×32=3072

A fully connected (feedforward) neural network accepts a one-dimensional feature vector as input. Therefore, each 3 × 32 × 32 image must be flattened into a single vector of length 3072 so that every input feature can connect to neurons in the first fully connected layer.

## Part II - Neural Networks

### Question 3: Helper functions

Complete two helpers (the intro-to-PyTorch lab is a good reference):

- `trainModel`: runs the training loop and returns the list of average epoch losses.
- `validate`: given a model and a data loader, returns the **accuracy, precision,
  recall, and F1-score**.

**Note:** code the metrics **manually** (counts of true/false positives and
negatives) - do not use scikit-learn or any other library for them. Treat
truck (label 1) as the positive class.

```python
def trainModel(model, n_epochs, optimizer, loss_fn, data_loader):
    """
    Train a model and return the list of average epoch losses.
    """
    model.train()
    epoch_losses = []

    for epoch in range(n_epochs):
        running_loss = 0.0

        for inputs, labels in data_loader:
            # Forward pass
            outputs = model(inputs)
            loss = loss_fn(outputs, labels)

            # Backpropagation
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        avg_loss = running_loss / len(data_loader)
        epoch_losses.append(avg_loss)

        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}")

    return epoch_losses


def validate(model, data_loader):
    """
    Return accuracy, precision, recall, and F1-score.
    Positive class = truck (label = 1).
    """
    model.eval()

    tp = fp = tn = fn = 0

    with torch.no_grad():
        for inputs, labels in data_loader:
            outputs = model(inputs)
            predictions = torch.argmax(outputs, dim=1)

            for pred, true in zip(predictions, labels):

                if pred == 1 and true == 1:
                    tp += 1
                elif pred == 1 and true == 0:
                    fp += 1
                elif pred == 0 and true == 0:
                    tn += 1
                elif pred == 0 and true == 1:
                    fn += 1

    accuracy = (tp + tn) / (tp + tn + fp + fn)

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0

    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    f1 = (2 * precision * recall / (precision + recall)
          if (precision + recall) > 0 else 0)

    return accuracy, precision, recall, f1
```

### Question 4: A single-hidden-layer feedforward network

Build a feedforward network and train it.

**Architecture:**
- Input: 3072 features
- Hidden layer: 256 nodes, ReLU activation
- Output: 2 logits (one per class)

**Compile / train:**
- Loss: cross-entropy
- Optimizer: Adam
- Batch size: 32
- At least 100 epochs

Plot the average epoch loss, then report accuracy, precision, recall, and F1 on
**both** the training and test sets.

```python
class FFN(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(3072, 256)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(256, 2)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
# Create model
model = FFN()

# Loss function
loss_fn = nn.CrossEntropyLoss()

# Optimizer
optimizer = optim.Adam(model.parameters())

# Train for at least 100 epochs
losses = trainModel(
    model=model,
    n_epochs=50,
    optimizer=optimizer,
    loss_fn=loss_fn,
    data_loader=train_loader
)
```

```python
plt.figure(figsize=(8,5))
plt.plot(losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Average Loss")
plt.grid(True)
plt.show()
```

```python
train_acc, train_prec, train_rec, train_f1 = validate(model, train_loader)
test_acc, test_prec, test_rec, test_f1 = validate(model, test_loader)

print("Training Set")
print(f"Accuracy : {train_acc:.4f}")
print(f"Precision: {train_prec:.4f}")
print(f"Recall   : {train_rec:.4f}")
print(f"F1-score : {train_f1:.4f}")

print("\nTest Set")
print(f"Accuracy : {test_acc:.4f}")
print(f"Precision: {test_prec:.4f}")
print(f"Recall   : {test_rec:.4f}")
print(f"F1-score : {test_f1:.4f}")
```

### Question 5: Your turn - build a better network

Design your own feedforward network, `NewNet`, aiming to **exceed an F1-score of
0.80** on the test set. You don't need formal cross-validation or hyperparameter
search - just experiment locally until you have an architecture that works.

Knobs you can turn (all from lecture):
- number and width of hidden layers
- activation functions (ReLU, tanh, sigmoid)
- dropout
- learning rate
- optimizer (SGD vs Adam) and weight decay
- number of epochs

(Feedforward only - no convolutions.) Train for at least 100 epochs, then report the
metrics on the train and test sets.

```python
class NewNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(3072, 512),
            nn.ReLU(),
            nn.Dropout(0.30),

            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.30),

            nn.Linear(256, 128),
            nn.ReLU(),

            nn.Linear(128, 2)
        )

    def forward(self, x):
        return self.network(x)
```

```python
# Create model
new_model = NewNet()

# Loss function
loss_fn = nn.CrossEntropyLoss()

# Adam optimizer with weight decay
optimizer = optim.Adam(
    new_model.parameters(),
    lr=0.001,
    weight_decay=1e-4
)

# Train for at least 100 epochs
losses = trainModel(
    model=new_model,
    n_epochs=50,
    optimizer=optimizer,
    loss_fn=loss_fn,
    data_loader=train_loader
)
```

```python
plt.figure(figsize=(8,5))
plt.plot(losses)
plt.title("NewNet Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Average Loss")
plt.grid(True)
plt.show()
```

```python
train_acc, train_prec, train_rec, train_f1 = validate(new_model, train_loader)
test_acc, test_prec, test_rec, test_f1 = validate(new_model, test_loader)

print("Training Set")
print(f"Accuracy : {train_acc:.4f}")
print(f"Precision: {train_prec:.4f}")
print(f"Recall   : {train_rec:.4f}")
print(f"F1-score : {train_f1:.4f}")

print("\nTest Set")
print(f"Accuracy : {test_acc:.4f}")
print(f"Precision: {test_prec:.4f}")
print(f"Recall   : {test_rec:.4f}")
print(f"F1-score : {test_f1:.4f}")
```

### Question 6: Overfitting and regularization

Show overfitting, then fix it.

- Train a **large** feedforward network (e.g. two wide hidden layers, no dropout) and
  record **both** train and test accuracy across epochs. You should see the train
  accuracy climb while the test accuracy stalls or drops - the train/test gap is
  overfitting.
- Now train a network of the **same capacity** but with a regularizer from lecture
  (dropout and/or weight decay). Show that the train/test gap shrinks.
- **Short answer:** what is overfitting, and why does your chosen fix help?

```python
class OverfitNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(3072, 1024),
            nn.ReLU(),

            nn.Linear(1024, 512),
            nn.ReLU(),

            nn.Linear(512, 2)
        )

    def forward(self, x):
        return self.network(x)
```

```python
def accuracy(model, data_loader):
    model.eval()

    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in data_loader:
            outputs = model(inputs)
            preds = torch.argmax(outputs, dim=1)

            correct += (preds == labels).sum().item()
            total += labels.size(0)

    return correct / total
```

```python
overfit_model = OverfitNet()

loss_fn = nn.CrossEntropyLoss()

optimizer = optim.Adam(overfit_model.parameters(), lr=0.001)

train_acc = []
test_acc = []

epochs = 50

for epoch in range(epochs):

    overfit_model.train()

    for x, y in train_loader:

        optimizer.zero_grad()

        outputs = overfit_model(x)

        loss = loss_fn(outputs, y)

        loss.backward()

        optimizer.step()

    train_acc.append(accuracy(overfit_model, train_loader))
    test_acc.append(accuracy(overfit_model, test_loader))
```

```python
class RegularizedNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(3072, 1024),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(512, 2)
        )

    def forward(self, x):
        return self.network(x)
```

```python
regularized_model = RegularizedNet()

optimizer = optim.Adam(
    regularized_model.parameters(),
    lr=0.001,
    weight_decay=1e-4
)

train_acc_reg = []
test_acc_reg = []

for epoch in range(epochs):

    regularized_model.train()

    for x, y in train_loader:

        optimizer.zero_grad()

        outputs = regularized_model(x)

        loss = loss_fn(outputs, y)

        loss.backward()

        optimizer.step()

    train_acc_reg.append(accuracy(regularized_model, train_loader))
    test_acc_reg.append(accuracy(regularized_model, test_loader))
```

```python
### YOUR CODE HERE
plt.figure(figsize=(10,6))

plt.plot(train_acc, label="Train (No Regularization)")
plt.plot(test_acc, label="Test (No Regularization)")

plt.plot(train_acc_reg, "--", label="Train (Regularized)")
plt.plot(test_acc_reg, "--", label="Test (Regularized)")

plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Effect of Regularization")
plt.legend()
plt.grid(True)

plt.show()
```

*Your short answer here.*

What is overfitting?

Overfitting occurs when a neural network learns the training data too closely, including noise and specific patterns that do not generalize to new data. As a result, the training accuracy continues to increase while the test accuracy stops improving or even decreases.

Why does regularization help?

Dropout randomly disables neurons during training, preventing the network from relying too heavily on specific features. Weight decay discourages very large weights by adding a penalty to the loss function. Together, these techniques reduce overfitting, leading to better generalization and a smaller gap between training and test accuracy.

### Extra credit (optional): SGD vs Adam

Train the single-hidden-layer `FFN` twice - once with `SGD` and once with `Adam`
(same learning rate) - and plot the two training-loss curves on the same axes.
Which converges faster here?

```python
# SGD
model_sgd = FFN()

loss_fn = nn.CrossEntropyLoss()

optimizer_sgd = optim.SGD(
    model_sgd.parameters(),
    lr=0.001
)

losses_sgd = trainModel(
    model=model_sgd,
    n_epochs=100,
    optimizer=optimizer_sgd,
    loss_fn=loss_fn,
    data_loader=train_loader
)


# Adam
model_adam = FFN()

optimizer_adam = optim.Adam(
    model_adam.parameters(),
    lr=0.001
)

losses_adam = trainModel(
    model=model_adam,
    n_epochs=100,
    optimizer=optimizer_adam,
    loss_fn=loss_fn,
    data_loader=train_loader
)
```

```python
### YOUR CODE HERE
plt.figure(figsize=(8,5))

plt.plot(losses_sgd, label="SGD")
plt.plot(losses_adam, label="Adam")

plt.xlabel("Epoch")
plt.ylabel("Average Training Loss")
plt.title("SGD vs Adam")
plt.legend()
plt.grid(True)

plt.show()
```
